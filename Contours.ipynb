```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('/content/drive/MyDrive/MRT/taj.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
imgray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
ret, thresh0 = cv2.threshold(imgray, 127, 255, 0)
ret1,thresh1 = cv2.threshold(imgray,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(imgray,127,255,cv2.THRESH_TRUNC)
ret,thresh3 = cv2.threshold(imgray,127,255,cv2.THRESH_TOZERO)
ret,thresh4 = cv2.threshold(imgray,127,255,cv2.THRESH_TOZERO_INV)
```
