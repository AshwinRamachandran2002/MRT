```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Gaussian Noise
Main sources- due to poor illumination, high temperature, transmission

```python
gaussian=cv2.imread('/content/drive/MyDrive/MRT/gauss.png')
plt.imshow(gaussian)
```

```python
avg_filter = cv2.blur(gaussian,(10,10))
plt.imshow(avg_filter)
```

```python
GaussianKernelStrip=cv2.getGaussianKernel(21,5)
GaussianKernel= cv2.sepFilter2D(gaussian,-1,GaussianKernelStrip,GaussianKernelStrip)
plt.imshow(GaussianKernel)
```

```python
mediannew = cv2.medianBlur(gaussian,5)
plt.imshow(mediannew)
```

```python
 bilateral = cv2.bilateralFilter(gaussian, 15, 95, 75)
 plt.imshow(bilateral)
```

SnP noise is removed best by median filtering. Caused due to improper analog conversion

```python
snp=cv2.imread('/content/drive/MyDrive/MRT/snp.jpeg')
plt.imshow(snp)
```

```python
mediannew = cv2.medianBlur(snp,5)
plt.imshow(mediannew)
```
