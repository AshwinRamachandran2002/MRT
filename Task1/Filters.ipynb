```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/drive/MyDrive/MRT/filtering.png')
plt.imshow(image)
```

Gaussian Filtering- A kernel following the gaussian distribution

```python
avg_filter = cv2.blur(image,(21,21))
plt.imshow(avg_filter)
```

```python
GaussianKernelStrip=cv2.getGaussianKernel(21,5)
GaussianKernel= cv2.sepFilter2D(image,-1,GaussianKernelStrip,GaussianKernelStrip)
```

```python
#c = cv2.GaussianBlur(image,(5,5),1)
plt.imshow(GaussianKernel)
```

Median Filtering for SnP noise. It takes the median value of the pixels in the kernel size

```python
median=cv2.imread('/content/drive/MyDrive/MRT/snp.jpeg')
plt.imshow(median)
```

```python
mediannew = cv2.medianBlur(median,5)
plt.imshow(mediannew)
```

Bilateral Filtering- Reducing the noise in the images and preserve the edges
It considers only the pixels within a d ditance and those that are within sigmaclor and sigmaspace values

```python
image2 = cv2.imread('/content/drive/MyDrive/MRT/2.png')
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
```

```python
 bilateral = cv2.bilateralFilter(image2, 15, 95, 75)
 #sigmacolor=95- the more the more color pixels are considered for filtering
 #sigmspace=75
 #distance =15
 plt.imshow(bilateral)
```

```python
image = cv2.circle(image2, (55,55), 35, (0,255,0), 5)
plt.imshow(image)
```

```python
image=cv2.line(image, (70,70), (100,10), (255,0,0), 2)
plt.imshow(image)
```

Sharpening Images


```python
sharp=cv2.imread('/content/drive/MyDrive/MRT/sharp.jpeg')
plt.imshow(sharp)
```

```python
filter=np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
sharpened=cv2.filter2D(sharp,-1,filter)
plt.imshow(sharpened)
```

```python
for i in range(5,20):
  filter[1][1]=i
  sharpened=cv2.filter2D(sharp,-1,filter)
  plt.figure()
  plt.imshow(sharpened)
```

```python
filter2=np.array([[0,0,-1,0,0],[0,-1,-2,-1,0],[-1,-2,16,-2,-1],[0,-1,-2,-1,0],[0,0,-1,0,0]])
sharpened=cv2.filter2D(sharp,-1,filter2)
plt.imshow(sharpened)
```
