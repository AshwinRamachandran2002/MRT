```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
list=[[-1,0,1],[-2,0,2],[-1,0,1]]
sobelFilter=np.asarray(list)
sobelFilter
```

```python
image2 = cv2.imread('/content/drive/MyDrive/MRT/3.jpg')
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)
plt.imshow(image2,cmap='gray')
```

```python
imagearr=np.asarray(image2)
```

```python
imagearr.shape
```

```python
horizz=imagearr.shape[1]-2
verizz=imagearr.shape[0]-2
no=0
verticalmak=np.zeros((verizz,horizz))
currarr=imagearr
for j in range(verticalmak.shape[1]):
  for k in range(verticalmak.shape[0]):
    if((currarr[k:k+3,j:j+3]).shape!=(3,3)):
      print(j,k)
      continue
    verticalmak[k][j]=np.sum(sobelFilter*currarr[k:k+3,j:j+3])
```

```python
plt.imshow(verticalmak,cmap='gray')
```

```python
list=[[-1,-2,-1],[0,0,0],[1,2,1]]
sobelFiltervert=np.asarray(list)
sobelFiltervert
```

```python
horizz=imagearr.shape[1]-2
verizz=imagearr.shape[0]-2
no=0
horizontalmak=np.zeros((verizz,horizz))
currarr=imagearr
for j in range(horizontalmak.shape[1]):
  for k in range(horizontalmak.shape[0]):
    if((currarr[k:k+3,j:j+3]).shape!=(3,3)):
      print(j,k)
      continue
    horizontalmak[k][j]=np.sum(sobelFiltervert*currarr[k:k+3,j:j+3])
```

```python
plt.imshow(horizontalmak,cmap='gray')
```
